```python
#Contains Base class Animal and it is inherited by mammal,bird,amphibian,reptile and fish classes
class Animal:
  def __init__(self, name, species, diet):
    self.name = name
    self.species = species
    self.diet = diet

  def get_info(self):
    return f"Name: {self.name}\nSpecies: {self.species}\nDiet: {self.diet}"

class Mammal(Animal):
  def __init__(self, name, species, diet, fur, description):
    super().__init__(name, species, diet)
    self.fur = fur
    self.description = description

  def get_info(self):
    return super().get_info() + f"\nFur: {self.fur}\nDescription: {self.description}"

class Reptile(Animal):
  def __init__(self, name, species, diet, scales, description):
    super().__init__(name, species, diet)
    self.scales = scales
    self.description = description

  def get_info(self):
    return super().get_info() + f"\nScales: {self.scales}\nDescription: {self.description}"

class Fish(Animal):
  def __init__(self, name, species, diet, fins, description):
    super().__init__(name, species, diet)
    self.fins = fins
    self.description = description

  def get_info(self):
    return super().get_info() + f"\nFins: {self.fins}\nDescription: {self.description}"

class Amphibian(Animal):
  def __init__(self, name, species, diet, habitat, description):
    super().__init__(name, species, diet)
    self.habitat = habitat
    self.description = description

  def get_info(self):
    return super().get_info() + f"\nHabitat: {self.habitat}\nDescription: {self.description}"

class Bird(Animal):
  def __init__(self, name, species, diet, feathers, description):
    super().__init__(name, species, diet)
    self.feathers = feathers
    self.description = description

  def get_info(self):
    return super().get_info() + f"\nFeathers: {self.feathers}\nDescription: {self.description}"
```
